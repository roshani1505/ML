## Q1. Write a Pandas program to add, subtract, multiple and divide two Pandas Series.

```python
s1 = pd.Series([2, 4, 6, 8, 10])
s2 = pd.Series([1, 3, 5, 7, 9])
print("Addition of Two Series:",s1+s2)
print("Subtration of Two Series:",s1-s2)
print("Multiplication of Two Series:",s1*s2)
print("Division of Two Series:",s1/s2)
```

## Q2. Write a Pandas program to compare (<,>,=) the elements of the two Pandas Series.

```python
s2 = pd.Series([1, 3, 5, 7, 10])
print("s1 is greater than s2:",s1>s2)
print("S1 is less than s2:",s1<s2)
print("Both are equal:",s1==s2)
```

## Q4. Write a Pandas program to convert Series of lists to one Series.

```python
sr1 = pd.Series([['Red', 'Green', 'White'],['Red', 'Black'],['Yellow']])
print("Original Series of list")
print(sr1)
print("One Series")
s = sr1.explode()
s.index = [0,1,2,3,4,5]
print(s)
```

## Q9. Write a Pandas program to get the day of month, day of year, week number and day of week from a given series of date strings.

```python
sr6 = pd.Series(['01 Jan 2015', '10-02-2026','20180307','2014/05/06','2016-04-12','2019-04-12:20'])
print("Original Series")
print(sr6)
dates = pd.to_datetime(sr6, errors='coerce')
result = pd.DataFrame({
    "Original Date": dates,
    "Day of Month": dates.dt.day,
    "Day of Year": dates.dt.dayofyear,
    "Week Number": dates.dt.isocalendar().week,
    "Day of Week": dates.dt.day_name()
})

print(result)
```

## 11. Write a Pandas program to get the first 3 rows of a given DataFrame.

```python
import numpy as np
exam_data = {
    'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin','Jonas'],
    'score': [12.5,9,16.5,'np.nan',9,20,14.5,'np.nan',8,19],
    'attempts': [1,3,2,3,2,3,1,1,2,1],
    'qualify': ['yes','no','yes','no','no','yes','yes','no','no','yes']
}
labels = ['a','b','c','d','e','f','g','h','i','j']
print("First three rows of the data frame:")
d6 = pd.DataFrame(exam_data, index= labels)
print(d6.head(3))
```

## 12. Write a Pandas program to select the rows where the number of attempts in the examination is greater than 2.(as per data set given in Q.11)

```python
print("Number of attempts in the examination is greater than 2:")
print(d6[d6['attempts'] > 2])
```

## 13. Write a Pandas program to select the rows where number of attempts in the examination is less than 2 and score greater than 15. (as per data set given in Q.11)

```python
d6['score'] = pd.to_numeric(d6['score'], errors='coerce')
result = d6[(d6['attempts'] < 2) & (d6['score'] > 15)]
print(result)
```

## 14. Write a Pandas program to calculate the sum of the examination attempts by the students. (as per data set given in Q.11)

```python
total_attempts = d6['attempts'].sum()
print("Total examination attempts by students:", total_attempts)
```

## 15. Write a Pandas program to sort the DataFrame first by 'name' in descending order, then by 'score' in ascending order.

```python
sorted_df = d6.sort_values(
    by=['name', 'score'],
    ascending=[False, True]
)

print(sorted_df)
```

## 17. Write a Pandas program to find the row for where the value of a given column is maximum.

```python
dt2 = {
    'col1': [1,2,3,4,7],
    'col2': [4,5,6,9,5],
    'col3': [7,8,12,1,11]
}
d8 = pd.DataFrame(dt2)
print("Original DataFrame")
print(d8)
print("Row where value of col1 is maximum:")
print(d8.loc[d8['col1'].idxmax()])
print("Row where value of col2 is maximum:")
print(d8.loc[d8['col2'].idxmax()])
print("Row where value of col3 is maximum:")
print(d8.loc[d8['col3'].idxmax()])
```
