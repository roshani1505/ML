## 6. Create a 2d array with 1 on the border and 0 inside. 

```python
c = np.ones((10,10))
```

```python
print(c)
```

```python
c[1:-1, 1:-1] = 0
print(c)
```

## 7. Create a 8x8 matrix and fill it with a checkerboard pattern.

```python
f = np.ones((8,8), dtype=int)
f[::2, ::2] = 0
f[1::2, 1::2] = 0
print(f)
```

## 8. Create a 5x5 matrix with values 1,2,3,4 just below the diagonal.

```python
m = np.diag([1, 2, 3, 4], k=-1)
print(m)
```

## 10. Create random vector of size 10 and replace the maximum value by 0.

```python
d = np.random.random((10))
print(d)
```

```python
max = d.max()
d[d==max] = 0
print(d)
```
