```python
numbers = []

for i in range(10):
    num = int(input("Enter a number: "))
    numbers.append(num)

t = tuple(numbers)

min_val = t[0]
max_val = t[0]

for i in range(len(t)):
    if t[i] < min_val:
        min_val = t[i]
    if t[i] > max_val:
        max_val = t[i]

print("Min =", min_val)
print("Max =", max_val)
```
