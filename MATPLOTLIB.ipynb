```python
subjects = ['Math', 'Physics', 'Chemistry', 'English', 'CS'] 
marks = [78, 85, 72, 80, 90] 
plt.bar(subjects, marks) 
plt.title('Marks vs Subjects') 
plt.xlabel('Subjects') 
plt.ylabel('Marks') 
plt.show() 
```

```python
months = ["Jan", "Feb", "Mar", "Apr"]   
visitors = [1200, 1800, 2100, 1600]   
plt.plot(months, visitors, color="blue",  marker="o",  linestyle="--")  
plt.title("Monthly Website Visitors")  
plt.xlabel("Month")                   
plt.ylabel("Visitors")   
plt.grid(True)
```

```python
products = ["Laptops", "Phones", "Tablets"]  # Categories 
sales = [200, 350, 150]                      
# Values 
plt.bar(products, sales, color=["#4CAF50", "#2196F3", "#FF5722"], width=0.6)                                
plt.title("Q1 Product Sales") 
plt.ylabel("Units Sold (Thousands)") 
plt.ylim(0, 400)   
# to add text on each  
for i, v in enumerate(sales): 
    plt.text(i, v, str(v), ha='center')
```
