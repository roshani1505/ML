## We use the dot function to compute inner products of vectors

```python
x = np.array([[1, 2], [3, 4]])
y = np.array([[5, 6], [7, 8]])

v = np.array([9, 10])
w = np.array([11, 12])
```

```python
# Inner product of vectors (cross product of matrix)
print(v.dot(w))
print(np.dot(v, w))
```

```python
print(x.dot(v))
print(np.dot(x, v))
```

```python
print(x.dot(y))
print(np.dot(x, y))
```
